```python
def ticket(age, gender):
  if age < 12:
    status = "child"
  elif age < 65 and gender == "male":
    status = "male adult"
  elif age < 65 and gender == "female":
    status = "female adult"
  else:
    status = "senior"
  return status

print(ticket(60, "female"))
```
